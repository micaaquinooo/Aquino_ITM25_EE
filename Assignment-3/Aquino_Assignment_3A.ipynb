# Birthmonth Cleaning

````python
import csv
with open('recweekdata.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    dirty_month_list = []
    for row in reader:
        dirty_month_list.append(row['birthmonth'])
    print(dirty_month_list)
````

````python
# Create a dictionary that assigns the MONTH to its ABBREVIATION

month_number = {'January':'1','February':'2','March':'3','April':'4','May':'5','June':'6',
          'July':'7','August':'8','September':'9','October':'10','November':'11','December':'12'}

month_short = {'January':'Jan','February':'Feb','March':'Mar','April':'Apr','May':'May','June':'Jun',
          'July':'Jul','August':'Aug','September':'Sep','October':'Oct','November':'Nov','December':'Dec'}

clean_month_list = []


# loop through list above to change all abbreviations
for i in dirty_month_list:
    
    # NUMBERS
    if len(i) <= 2:
        for j in month_number:
            if i == month_number[j]:
                clean_month_list.append(j)
                
    # 3 LETTER ABBRV
    elif len(i) == 3:
        for k in month_short:
            if i == month_short[k]:
                clean_month_list.append(k)
                
    else:
        clean_month_list.append(i)
                
clean_month_list        
        
````

````python
month_tally = {}

for i in clean_month_list:
    month_tally.setdefault(i,clean_month_list.count(i))

month_tally
````

# Birthyear Cleaning

````python
import csv
with open('recweekdata.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    dirty_year_list = []
    for row in reader:
        dirty_year_list.append(row['birthyear'])
    print(dirty_year_list)
````

````python
clean_year_list = []
for i in dirty_year_list:
    
    # extract years with 4 DIGITS
    if len(i) == 2:
        clean_year_list.append(i)
    
    # CONCATENATE LAST 2 DIGITS
    else:
        clean_i = i[2]+i[3]
        clean_year_list.append(clean_i)
    
print(clean_year_list)
````

## Number Cleaning

````python
import csv
with open('recweekdata.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    dirty_number_list = []
    for row in reader:
        dirty_number_list.append(row['phonenumber'])
    print(dirty_number_list)
````

````python
clean_number_list = []
for i in dirty_number_list:
    
    # REMOVE SPACES
    i = i.replace(" ","")
    
    # APPEND NUMBERS THAT ARE IN (+63) FORMAT
    if len(i) == 13:
        clean_number_list.append(i)
        
    # CONCATENATE (+63) TO REMAINING NUMBERS
    else:
        new_i = "+63"+i
        clean_number_list.append(new_i)
print(clean_number_list)
````

````python
globe_list = []
smart_list = []
for i in clean_number_list:
    # SMART = 0918
    # test 4th and 5th number
    if i[4] == '1' and i[5] == '8':
        smart_list.append(i)
        
    # GLOBE = everything else
    else:
        globe_list.append(i)
  
````

````python
globe_count = 0
smart_count = 0
for i in globe_list:
    globe_count += 1
    
for i in smart_list:
    smart_count += 1

number_tally = {'Globe':globe_count,'Smart':smart_count}
number_tally
````

````python
for i in sorted (number_tally):
    print(i,number_tally[i])
````
