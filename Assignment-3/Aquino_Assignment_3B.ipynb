# Month Cleaning

````python
months = {"January":["1","Jan","January"],
          "February":["2","Feb","February"],
          "March":["3","Mar","March"],
          "April":["4","Apr","April"],
          "May":["5","May"],
          "June":["6","Jun","June"],
          "July":["7","Jul","July"],
          "August":["8","Aug","August"],
          "September":["9","Sep","September"],
          "October":["10","Oct","October"],
          "November":["11","Nov","November"],
          "December":["12","Dec","December"]}
````

````python
def fix_month(month):
    for i in months:
        if month in months[i]:
            return i
fix_month("11")
````

````python
import re
recweek.loc[:,"birthmonth"] = recweek.loc[:,"birthmonth"].apply(lambda x: fix_month(x))
recweek
````

# Year Cleaning

````python
# change birthyear type from integer to string
recweek.loc[:,"birthyear"]=recweek.loc[:,"birthyear"].apply(str)
recweek.dtypes
````

````python
recweek.loc[:,"birthyear"]= recweek.loc[:,"birthyear"].str.replace("^99$", "1999",regex=True).copy()
recweek.loc[:,"birthyear"]= recweek.loc[:,"birthyear"].str.replace("^98$", "1998",regex=True).copy()
recweek.loc[:,"birthyear"]= recweek.loc[:,"birthyear"].str.replace("^97$", "1997",regex=True).copy()
recweek.loc[:,"birthyear"]= recweek.loc[:,"birthyear"].str.replace("^00$", "2000",regex=True).copy()
recweek.loc[:,"birthyear"]= recweek.loc[:,"birthyear"].str.replace("^01$", "2001",regex=True).copy()

recweek
````

# Course Cleaning

````python
courses = {"BS Communications Technology":["CTM","BS CTM","Communications Technology"],
           "BS Management Engineering":["ME","BS ME","Management Engineering"],
           "BS Management":["MGT","BS MGT","Management"],
           "BS Legal Management":["LM","BS LM","Legal Management"],
           "BS IT Entrepreneurship":["ITE","BS ITE","Information Technology Entrepreneurship"]}
````

````python
def fix_course(course):
    for i in courses:
        if course in courses[i]:
            return i
fix_course("ME")
````

````python
recweek.loc[:,"course"] = recweek.loc[:,"course"].apply(lambda x: fix_course(x))
recweek
````

# Number Cleaning

````python
recweek.loc[:,"phonenumber"]= recweek.loc[:,"phonenumber"].str.replace("^\+63 ", "",regex=True).copy()
recweek.loc[:,"phonenumber"]= recweek.loc[:,"phonenumber"].str.replace("^9", "09",regex=True).copy()

recweek
````

# Same Month

````python
final_recweek.groupby(["birthmonth"])["firstname"].count()
````

````python
month_count = final_recweek.groupby(["birthmonth"])["firstname"].count()
month_count = month_count.to_frame(name="count")
month_count.reset_index(inplace=True)
month_count
````

````python
month_dict = {"January":1,
              "February":2,
              "March":3,
              "April":4,
              "May":5,
              "June":6,
              "July":7,
              "August":8,
              "September":9,
              "October":10,
              "November":11,
              "December":12}
````

````python
month_fix = month_count.loc[:,"birthmonth"].apply(lambda x: re.sub(x,str(month_dict[x]), x) )
month_fix = month_fix.to_frame(name="new_index")
month_fix = month_fix.loc[:,"new_index"].apply(pd.to_numeric)
````

````python
month_count = month_count.join(month_fix)
month_count = month_count.set_index("new_index")
month_count
````

````python
month_count = month_count.sort_index()
month_count
````

# Same Day

````python
day_count = final_recweek.groupby(["birthday"])["firstname"].count()
day_count = day_count.to_frame(name="count")
day_count.reset_index(inplace=True)
day_count
````

# Same Year

````python
year_count = final_recweek.groupby(["birthyear"])["firstname"].count()
year_count = year_count.to_frame(name="count")
year_count.reset_index(inplace=True)
year_count
````

# Same Course

````python
course_count = final_recweek.groupby(["course"])["firstname"].count()
course_count = course_count.to_frame(name="count")
course_count.reset_index(inplace=True)
course_count
````

# Globe or Smart

````python
carrier = {"0917":"Globe", "0927":"Globe", "0918":"Smart"}
````

````python
final_recweek.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])
````

````python
final_recweek["carrier"] = final_recweek.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])
final_recweek
````

````python
phone_count = final_recweek.groupby(["carrier","gender"])["firstname"].count()
phone_count
````

````python
globe_female = phone_count[("Globe","Female")]
globe_male = phone_count[("Globe","Male")]
smart_female = phone_count[("Smart","Female")]
smart_male = phone_count[("Smart","Male")]
````

````python
gender_count = phone_count.to_frame(name="count")
gender_count.reset_index(inplace=True)
gender_count
````

````python
carrier_count = gender_count.loc[:,["carrier","count"]]
carrier_count = carrier_count.set_index("carrier").sum(level="carrier")
carrier_count.reset_index(inplace=True)
carrier_count
````

# Summary Stats

````python
from IPython.core.display import HTML

html = """
<html>
    <body>
         <h2>Summary Statistics</h2>
         <table>
             <tr><th>Stat</th><th>Value</th></tr>
             <tr><td>People born in each month</td><td>{month_count} </td></tr>
             <tr><td>People born on each day</td><td>{day_count} </td></tr>
             <tr><td>People born in each year</td><td>{year_count} </td></tr>
             <tr><td>People coming from each course</td><td>{course_count} </td></tr>
             <tr><td>People having numbers from Globe and Smart</td><td>{carrier_count}</td></tr>
             <tr><td>Male Globe subscribers:</td><td>{globe_male}</td></tr>
             <tr><td>Female Globe subscribers:</td><td>{globe_female}</td></tr>
             <tr><td>Male Smart subscribers:</td><td>{smart_male}</td></tr>
             <tr><td>Female Smart subscribers:</td><td>{smart_female}</td></tr>
         </table>
    </body>
</html>
"""

html_output = html.format(month_count = month_count,
                          day_count = day_count,
                          year_count = year_count,
                          course_count = course_count,
                          carrier_count = carrier_count,
                          globe_male = globe_male,
                          globe_female = globe_female,
                          smart_male = smart_male,
                          smart_female = smart_female                          
                         )

HTML(html_output)
````
